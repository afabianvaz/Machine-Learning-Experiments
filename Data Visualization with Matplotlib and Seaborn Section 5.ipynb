# Data Visualization with Matplotlib and Seaborn

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as ans
```

```python
titanic = pd.read_csv("titanic.csv")
titanic.head(5)
```

```python
titanic.info()
```

```python
titanic.plot(subplots = True, figsize = (15,12), sharex= False, sharey = False)
plt.show()
```

```python
titanic.age.plot(figsize = (12,8), sharex= False, sharey = False)
plt.show()
```

# Customization of Plots

```python
titanic.age.plot(figsize = (12,5), fontsize = 13, c ="r", linestyle = "-", xlim=(0,900), ylim=(0,80), 
                 xticks=xticks, yticks=yticks, rot = 90)
plt.title('titanic - Ages', fontsize = 15)
plt.legend(loc = 5, fontsize =15)
plt.xlabel("Passenger No", fontsize = 15)
plt.ylabel('Age', fontsize = 13)
plt.grid()
plt.show()
```

# Histogramms (Part 1)

```python
plt.style.use('seaborn')
```

```python
titanic.head()
```

```python
titanic.age.value_counts(bins = 10)
```

```python
titanic.age.plot(kind= "hist", figsize = (12,7), fontsize = 1, bins = 50)
plt.show()
```

```python
titanic.age.hist(figsize = (12,8), xlabelsize = 15, ylabelsize =15, bins = 80)
plt.show()
```

```python
plt.figure(figsize = (12,8))
plt.hist(titanic.age, density = False, cumulative = False, bins = 50)
plt.show()
```

```python
plt.figure(figsize = (12,8))
plt.hist(titanic.age.dropna(), density = True, cumulative = True, bins = 50)
plt.show()
```
