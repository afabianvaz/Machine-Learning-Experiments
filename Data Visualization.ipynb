# Data Visualization

```python
import matplotlib.pyplot as plt 
import numpy as np
import pandas as pd
%matplotlib inline

import mpl_toolkits
```

```python
data= open('agedata2.csv','r')
agefiles =data.readlines()

city_list = []

for records in agefiles:
    age, city = records.split(sep= ',')
    city_list.append(city)
```

```python
from collections import Counter
city_count = Counter(city_list)

city_names = list(city_count.keys())
city_values = list(city_count.values())

plt.pie(city_values, labels=city_names, autopct = '%.2f%%')
plt.show()
```

```python
f= open('salesdata2.csv','r')
salefile = f.readlines()
```

```python
# Creat the sale list

s_list = []
c_list = []

#Append all the records from the files to the saleslist

for records in salefile:
    sale, cost = records.split(sep = ',')
    s_list.append(int(sale))
    c_list.append(int(cost))
```

```python
plt.title('Sales vs Cost')
plt.xlabel('Sale')
plt.ylabel('Cost')

plt.scatter(s_list, c_list, marker='*', s=100, color='#FF5733')
plt.show()
```
